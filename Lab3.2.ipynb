```python
'''
1. Given customer bank saving data customer-savings.txt, please answer the
following questions using Python programming.
1. What’s the average balance for male and female?
2. What’s the average balance for white collar and blue collar in England?
Data columns in customer-savings.txt (separated by comma):
Customer ID, Customer Name, Customer Surname, Gender, Age, Region, Job
Classification, Date joined, Balance
100000001,Simon,Walsh,Male,21,England,White Collar,05.Jan.15,113810.15
100000003,Liam,Brown,Male,46,England,White Collar,07.Jan.15,101536.83
'''

fh = open('customer-savings.txt', 'r')
countM = 0
countF = 0
totalMaleBalance = 0
totalFemaleBalance = 0

lines = fh.readlines()
myList = []

for line in lines:
    myList.append(line.split(','))


for i in myList:
    if(i[3] == 'Male'):
        countM += 1
        balance = i[-1]
        balance = balance.replace("\n", "")
        totalMaleBalance += float(balance)
    else:
        countF += 1
        balance = i[-1]
        balance = balance.replace("\n", "")
        totalFemaleBalance += float(balance)
        
print("Average Balance for Male :", (totalMaleBalance/countM))
print("Average Balance for Female :", (totalFemaleBalance/countF))


white = 0
blue = 0
whiteBalance = 0
blueBalance = 0

for i in myList:
    if(i[5]=='England'):
        if(i[6] == 'White Collar'):
            white += 1
            balance = i[-1]
            balance = balance.replace("\n", "")
            whiteBalance += float(balance)
        elif(i[6] == 'Blue Collar'):
            blue += 1
            balance = i[-1]
            balance = balance.replace("\n", "")
            blueBalance += float(balance)
print("\n")        
print("Average Balance for White Collar in England :", (whiteBalance/white))
print("Average Balance for Blue Collar in England :", (blueBalance/blue))
```
