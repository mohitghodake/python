```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("dengue_features_train.csv")
df1 = pd.read_csv("dengue_labels_train.csv")

df_final = pd.merge(df, df1)
df_final.head()
```

```python
df_final=df_final.fillna(df_final.mean())
df_final=df_final.drop(['week_start_date'], axis=1)
df_final['city']=df_final['city'].astype('category')
df_final=df_final.apply(pd.to_numeric, errors='coerce')
df_final.info()
```

```python
from sklearn.model_selection import train_test_split

X=df_final.loc[:, 'city':'station_precip_mm']
y=df_final.total_cases

#X=X.apply(pd.to_numeric, errors='coerce')
#y=y.apply(pd.to_numeric, errors='coerce')

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

clf=LinearRegression()
model=clf.fit(X_train, y_train)
```
